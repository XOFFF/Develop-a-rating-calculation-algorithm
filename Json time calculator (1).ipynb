```python
import json 
  
# Opening JSON file 
f = open('C:\\Users\\marke\\Desktop\\cache2.json',) 
  
# returns JSON object as a dictionary 
dataFromJSON = json.load(f) 
  
# Iterating through the json list 

timeList = []

employee_id = 0
name = ''
room_id = 0

for i in dataFromJSON: 
    timeList.append(i['Date time'])
    employee_id = i['id']
    name = i['name']
    room_id = i['Room id']
    print(i['Date time']) 

# Closing file 
f.close() 
```

## Finding difference between two events

```python
first_time = datetime.datetime.strptime('18:34:35.074991', '%H:%M:%S.%f')
later_time = datetime.datetime.strptime('19:32:35.074992', '%H:%M:%S.%f')
difference = later_time - first_time
print(difference)
```
