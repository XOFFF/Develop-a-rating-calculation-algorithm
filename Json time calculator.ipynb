```python
import json 
  
# Opening JSON file 
f = open('cache2.json',) 
  
# returns JSON object as a dictionary 
data = json.load(f) 
  
# Iterating through the json list 

timeList = []

for i in data: 
    timeList.append(i['Date time'])
    print(i['Date time']) 

# Closing file 
f.close() 
```

## Converting from str to obj

```python
import datetime

timeT = datetime.datetime.now().time()
print(timeT)
datetime_object = datetime.datetime.strptime('18:34:35.074991', '%H:%M:%S.%f').time()
print(datetime_object)
```

## Converting from obj to str

```python
timestampStr = timeT.strftime('%H:%M:%S.%f')
print(timestampStr)
```

## Finding difference between two events

```python
first_time = datetime.datetime.strptime('18:34:35.074991', '%H:%M:%S.%f')
later_time = datetime.datetime.strptime('19:32:35.074991', '%H:%M:%S.%f')
difference = later_time - first_time
print(difference)
```

```python
first_time = datetime.datetime.strptime('2020-05-10 20:11:39.411775', '%Y-%m-%d %H:%M:%S.%f')
later_time = datetime.datetime.now()
difference = later_time - first_time
print(difference)
```
